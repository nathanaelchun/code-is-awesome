```python
import pandas as pd 
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_formats = ['retina']
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df
```

```python
df.Outcome.value_counts()
```

```python
df.loc[:, ["Pregnancies", "Outcome"]]
```

```python
fig, ax = plt.subplots(figsize=(6,6), facecolor="white")

sns.stripplot(x=df.Outcome, y=df.Pregnancies, ax=ax)
ax.set_xlabel("Outcome")
ax.set_ylabel("Number of Pregnancies")
ax.set_xticklabels(["No","Yes"])
ax.set_yticks(range(0, df.Pregnancies.max() + 2))
ax.set_title("Diabetes vs. Pregnancies")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,6), facecolor="white")

sns.stripplot(x=df.Outcome, y=df.Age, ax=ax)
ax.set_xlabel("Outcome")
ax.set_ylabel("Age")
ax.set_xticklabels(["No","Yes"])
ax.set_title("Diabetes vs. Age")

plt.show()
```

```python
fig, axs = plt.subplots(nrows=4, ncols = 2,figsize=(12,24), facecolor="white")

axs = np.ravel(axs)
for i, col in enumerate(df.columns):
    
    if col == "Outcome":
        continue
    sns.stripplot(x=df.Outcome, y=df[col], ax=axs[i])
    
    axs[i].set_xlabel("Outcome")
    axs[i].set_ylabel(col)
    axs[i].set_xticklabels(["No","Yes"])
    axs[i].set_title(f"Diabetes vs. {col}")
plt.tight_layout()
plt.show()
    
```
