```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn import datasets, svm, metrics
from sklearn.model_selection import train_test_split
```

```python
digits = datasets.load_digits()
```

```python
digits
```

```python
n_samples = len(digits.images)
```

```python
n_samples
```

```python
digits.images
```

```python
index = 316
```

```python
digits.images[index]
```

```python
digits.images[index].shape
```

```python
digits.target[index]
```

```python
fig, ax = plt.subplots(figsize=(2, 2))

ax.imshow(digits.images[index])
ax.set_title("Label:" + str(digits.target[index]))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(2, 2))

ax.imshow(digits.images[index], cmap=plt.cm.gray_r, interpolation='nearest')
ax.set_title("Label:" + str(digits.target[index]))

plt.show()
```

```python
fig, axs = plt.subplots(nrows= 10, ncols = 10, figsize=(15, 15))
axs = np.ravel(axs)
for i in range(100):
    axs[i].imshow(digits.images[i], cmap=plt.cm.gray_r, interpolation='nearest')
    axs[i].set_title("Label:" + str(digits.target[i]))

plt.tight_layout()
plt.show()
```

```python
data = digits.images.reshape((n_samples, -1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
        data, digits.target, test_size=0.5, shuffle=False, random_state=0
)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
classifier = svm.SVC(gamma = 0.001, random_state=0)
```

```python
classifier.fit(X_train, y_train)
```

```python
predicted = classifier.predict(X_test)
```

```python
predicted
```

```python
fig, ax = plt.subplots(figsize=(2, 2))

index = 123

ax.imshow(X_test[index].reshape(8, -1))
ax.set_title("Prediction:" + str(predicted[index]))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(2, 2))

index = 53

ax.imshow(X_test[index].reshape(8, -1))
ax.set_title("Prediction:" + str(predicted[index]))

plt.show()
```

```python
fig, axs = plt.subplots(nrows= 10, ncols = 10, figsize=(15, 15))
axs = np.ravel(axs)
for i in range(100):
    axs[i].imshow(X_test[i].reshape(8, -1), cmap=plt.cm.gray_r, interpolation='nearest')
    axs[i].set_title("Label:" + str(predicted[i]))

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

metrics.plot_confusion_matrix(classifier, X_test, y_test, ax=ax)

plt.show()
```

```python
print(metrics.classification_report(y_test, predicted))
```
